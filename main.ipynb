
#
# Normalisation des données
#


```python
# Ajout d'un index

data_frame_nb_of_travelers.set_index("month")
```

```python
# On normalise avec la méthode min max

min_value = data_frame_nb_of_travelers['passengers'].min()
max_value = data_frame_nb_of_travelers['passengers'].max()
data_frame_nb_of_travelers['passengers'] = (data_frame_nb_of_travelers['passengers'] - min_value) / (max_value - min_value)
```
